```python
import pandas as pd
from sklearn import svm
```

```python
df = pd.read_csv('/Users/arman/Skripsi/Dataset/BigData/Fix/Piala-Dunia-Label-Clean.csv')
df = df.drop(labels="Unnamed: 0", axis=1)
df
```

```python
pd.value_counts(df['Label'])
```

## Pembobotan TF-IDF

```python
# TF-IDF
train_X = df['Steming']
train_Y = df['Label'] 

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer()
tfidf_vect.fit(train_X)
train_X_tfidf = tfidf_vect.transform(train_X)
```

```python
print(train_X_tfidf.shape)
```

```python
# You can use the below syntax to see the vocabulary that it has learned from the corpus
print(tfidf_vect.vocabulary_)
```

## Penerapan Model SVM

```python
from sklearn.svm import SVC

#Menerapkan pemodelan SVM pada data train

model = SVC(kernel='rbf', 
            decision_function_shape='ovr')
clf = model.fit(train_X_tfidf, train_Y)



# SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
#     decision_function_shape='ovr', degree=3, gamma='auto_deprecated',
#     kernel='linear', max_iter=-1, probability=False, random_state=None,
#     shrinking=True, tol=0.001, verbose=False)
```

```python
dp = pd.read_csv("/Users/arman/Skripsi/Dataset/Testing/coba.csv")
dp.head()
```

```python
# analisis sentimen use ur csv file
transform_tfidf =  tfidf_vect.transform(dp['Text'])
predict = clf.predict(transform_tfidf)

dp['Label-SVM'] = predict
dp.head()
```
